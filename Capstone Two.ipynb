```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\nmsmi\\Downloads\\archive (4)\\gendergapinaverage new.csv")
```

```python
sns.countplot(data=df,x='Year')
plt.xlabel('')
plt.ylabel('Gender wage gap %')
plt.title('Gender Wage Gap Globally')
sns.despine()
plt.show()
```

```python
sns.swarmplot(data=df,
            x='Year',
            y='Gender wage gap %')
plt.xlabel('Year')
plt.ylabel('Gender wage gap')
sns.despine()
plt.savefig('Gender breakdown per year')
plt.show()
```
